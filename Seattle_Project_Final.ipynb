```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
mpl.style.use('ggplot')
```

```python
seattle_collisions = 'https://s3.us.cloud-object-storage.appdomain.cloud/cf-courses-data/CognitiveClass/DP0701EN/version-2/Data-Collisions.csv'
df_collisionsorg = pd.read_csv(seattle_collisions, low_memory=False)
df_collisionsorg = df_collisionsorg[pd.to_datetime(df_collisionsorg.INCDATE).dt.year >= 2010]
df_collisionsorg = df_collisionsorg[df_collisionsorg['SEVERITYCODE']==2]
df_collisionsorg.dropna(subset = ["ADDRTYPE"], inplace=True)
df_collisionsorg
```

```python
df_collisionsorg1 = df_collisionsorg.drop(['X','Y','PEDCOUNT','OBJECTID','SEVERITYDESC', 'INCKEY', 'COLDETKEY', 'STATUS', 'INTKEY', 'LOCATION', 'EXCEPTRSNCODE', 'EXCEPTRSNDESC', 'PEDCYLCOUNT', 'PEDROWNOTGRNT', 'SDOTCOLNUM', 'ST_COLCODE', 'ST_COLDESC', 'SEGLANEKEY', 'CROSSWALKKEY', 'HITPARKEDCAR', 'SEVERITYCODE.1','INCDATE','INCDTTM','JUNCTIONTYPE', 'SDOT_COLCODE', 'UNDERINFL', 'REPORTNO', 'SDOT_COLDESC'], axis=1)
df_collisionsorg1
```

```python
df_collisionsorg1['LIGHTCOND'] = df_collisionsorg1.LIGHTCOND.replace('Dark - No Street Lights', 'Night')
df_collisionsorg1['LIGHTCOND'] = df_collisionsorg1.LIGHTCOND.replace('Dark - Street Lights Off', 'Night')
df_collisionsorg1['LIGHTCOND'] = df_collisionsorg1.LIGHTCOND.replace('Dark - Street Lights On', 'Night')
df_collisionsorg1['LIGHTCOND'] = df_collisionsorg1.LIGHTCOND.replace('Dawn', 'Day')
df_collisionsorg1['LIGHTCOND'] = df_collisionsorg1.LIGHTCOND.replace('Daylight', 'Day')
df_collisionsorg1['LIGHTCOND'] = df_collisionsorg1.LIGHTCOND.replace('Dusk', 'Day')
```

```python
!conda install -c conda-forge folium=0.5.0 --yes
import folium
seattle_map = folium.Map(location=[47.6219, -122.3517], zoom_start=10)
seattle_map
```

```python
df_collisions = df_collisionsorg[:300]
df_collisions
```

```python
df_collisions.dropna(subset = ["X"], inplace=True)
df_collisions.dropna(subset = ["Y"], inplace=True)
df_collisions
```

```python
accidents = folium.map.FeatureGroup()

for lat, lng, in zip(df_collisions.Y, df_collisions.X):
    accidents.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, 
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )


seattle_map.add_child(accidents)
```

```python
latitudes = list(df_collisions.Y)
longitudes = list(df_collisions.X)
labels = list(df_collisions.ADDRTYPE)

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(seattle_map)    
    

seattle_map.add_child(accidents)
```
